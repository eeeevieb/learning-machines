```python
import pandas as pd
```

```python
data = pd.read_csv('results_sim_task_0.csv')
data.head()
```

```python
run0 = data.loc[data.run == 0]
run0.head()
```

```python
import matplotlib.pyplot as plt
```

```python
# plt.plot(range(len(run0.frontC)), run0.frontRR.values, label='frontRR')
plt.plot(range(len(run0.frontC)), run0.frontR.values, label='frontR')
# plt.plot(range(len(run0.frontC)), run0.frontC.values, label='frontC')
# plt.plot(range(len(run0.frontC)), run0.frontL.values, label='frontL')
# plt.plot(range(len(run0.frontC)), run0.frontLL.values, label='frontLL')
# plt.plot(range(len(run0.frontC)), run0.backR.values, label='backR')
# plt.plot(range(len(run0.frontC)), run0.backC.values, label='backC')
plt.plot(range(len(run0.frontC)), run0.backL.values, label='backL')
plt.legend()
plt.show()
```

```python
for run in range(10):
    curr_run = data.loc[data.run == run]
    plt.plot(range(len(curr_run.frontC)), curr_run.frontRR.values, label='frontRR')
    plt.plot(range(len(curr_run.frontC)), curr_run.frontR.values, label='frontR')
    plt.plot(range(len(curr_run.frontC)), curr_run.frontC.values, label='frontC')
    plt.plot(range(len(curr_run.frontC)), curr_run.frontL.values, label='frontL')
    plt.plot(range(len(curr_run.frontC)), curr_run.frontLL.values, label='frontLL')
    plt.plot(range(len(curr_run.frontC)), curr_run.backR.values, label='backR')
    plt.plot(range(len(curr_run.frontC)), curr_run.backC.values, label='backC')
    plt.plot(range(len(curr_run.frontC)), curr_run.backL.values, label='backL')
    plt.title(f'Run {run}, simulation')
    plt.legend()
    plt.show()
```
